## Librairies

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

## Table de hachage

```python
class TableDeHachage:
    
    def __init__(self, N=0, f=None):
        
        self.taille = N
        self.fct_hachage = f
        self.table = [None]*N
        self.nb_collisions = 0
        
    def inserer(self, c):
        
        i, collision = 0, True
        
        while collision and i < self.taille:
            
            indice = self.fct_hachage(c, i)
            
            if self.table[indice] == None: # case libre
                self.table[indice] = c
                collision = False
            else:                          # case occupée
                i += 1                     # incrémenter tentative
                
        self.nb_collisions += i
```

## Expérience

```python
N = 1023

def f1(c, i):
    return (c + i) % N
```

```python
def experience(f):

    t = TableDeHachage(N=N, f=f)

    nb_collisions_l = []

    for _ in range(N // 2):

        cle = random.randint(1, 10000)
        t.inserer(cle)
        nb_collisions_l.append(t.nb_collisions)
        
    return nb_collisions_l
```

```python
nb_collisions_l = experience(f1)
```

```python
X = range(len(nb_collisions_l))
Y = nb_collisions_l

plt.plot(X, Y)
plt.xlabel("nb d'éléments insérés")
plt.ylabel("nb cumulé de collisions")
plt.show()
```

## 50 expériences: fonction de hachage linéaire

```python
nb_collisions_ll = []

for exp in range(50):
    
    nb_collisions_l = experience(f1)
    nb_collisions_ll.append(nb_collisions_l)
```

```python
nb_collisions_moyen = np.array(nb_collisions_ll)
nb_collisions_moyen = np.mean(nb_collisions_moyen, axis=0)
```

```python
# nb_collisions_moyen
```

```python
X = range(len(nb_collisions_moyen))
Y = nb_collisions_moyen

plt.plot(X, Y, label="fonction de hachage linéaire")
plt.xlabel("nb d'éléments insérés")
plt.ylabel("nb cumulé de collisions moyen (50 expériences)")
plt.legend()
plt.show()
```

## 50 expériences: fonction de hachage quadratique

```python
N = 1023

def f2(c, i):
    return (c + 3*i**2 + i) % N
```

```python
nb_collisions_2_ll = []

for exp in range(50):
    
    nb_collisions_l = experience(f2)
    nb_collisions_2_ll.append(nb_collisions_l)
```

```python
nb_collisions_moyen_2 = np.array(nb_collisions_2_ll)
nb_collisions_moyen_2 = np.mean(nb_collisions_moyen_2, axis=0)
```

```python
# nb_collisions_moyen
```

```python
X = range(len(nb_collisions_moyen_2))
Y = nb_collisions_moyen_2

plt.plot(X, Y, label="fonction de hachage quadraqtique")
plt.xlabel("nb d'éléments insérés")
plt.ylabel("nb cumulé de collisions moyen (50 expériences)")
plt.legend()
plt.show()
```

## Fonction de hachage linéaire vs quadratique

```python
X1 = range(len(nb_collisions_moyen))
Y1 = nb_collisions_moyen

X2 = range(len(nb_collisions_moyen_2))
Y2 = nb_collisions_moyen_2

plt.plot(X1, Y1, label="fonction de hachage linéaire")
plt.plot(X2, Y2, label="fonction de hachage quadraqtique")
plt.xlabel("nb d'éléments insérés")
plt.ylabel("nb cumulé de collisions moyen (50 expériences)")
plt.legend()
plt.show()
```
